```python
import bioread
import matplotlib.pyplot as plt

def load_visualise(data_file):
    '''
    loads data and annotations, plots them 
    and returns data and annotations
    '''
    # Load the data
    data = bioread.read_file(data_file)
    
    # Filter out the 'Muscle A - EMG100C' channel
    channels_to_exclude = ['Muscle A - EMG100C', 'DTU100C - Trigger View, UIM100C - A 1']
    filtered_channels = [channel for channel in data.channels if channel.name not in channels_to_exclude]
    
    # Explore signal
    plt.figure(figsize=(12, 6))
    for channel in filtered_channels:
        plt.plot(channel.time_index, channel.data, label=channel.name)
    plt.legend()
    plt.xlabel('Time (s)')
    plt.ylabel('Signal')
    plt.title('Filtered Channels Data')
    plt.show()

    # Determine a suitable zoom-in range
    # For demonstration, we will use the first 10% of the data to find the range
    zoom_start = 0
    zoom_end = int(0.1 * len(filtered_channels[0].time_index))
    
    # Zoom in
    plt.figure(figsize=(12, 3))
    for channel in filtered_channels:
        plt.plot(channel.time_index, channel.data, label=channel.name)
    plt.xlim(filtered_channels[0].time_index[zoom_start], filtered_channels[0].time_index[zoom_end])
    plt.legend()
    plt.xlabel('Time (s)')
    plt.ylabel('Signal')
    plt.title('Zoomed In View')
    plt.show()
    
    return data

# Specify the file path
data_file_path = '/Users/cockraml/Library/CloudStorage/OneDrive-QueenslandUniversityofTechnology/Desktop/EDA and ECG Analysis/Gerber acq files/gerber_example_1.acq'

# Call the function with your file path
data = load_visualise(data_file_path)
```
