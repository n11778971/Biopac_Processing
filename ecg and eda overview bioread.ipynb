```python
%pylab inline
import bioread
pylab.rcParams['figure.figsize'] = (14.0, 12.0)
```

```python
data = bioread.read_file('/Users/cockraml/Library/CloudStorage/OneDrive-QueenslandUniversityofTechnology/Desktop/EDA and ECG Analysis/Gerber acq files/gerber_example_1.acq')
```

```python
# Create a list of channels to exclude
channels_to_exclude = ['Muscle A - EMG100C', 'DTU100C - Trigger View, UIM100C - A 1']

# Filter the channels
filtered_channels = [channel for channel in data.channels if channel.name not in channels_to_exclude]
```

```python
data.channels
```

```python
import matplotlib.pyplot as plt

# Initialize the plot
plt.figure(figsize=(10, 6))

# Plot each channel's data
for channel in filtered_channels:
    plt.plot(channel.time_index, channel.data, label=channel.name)

# Add legend and labels
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Signal')
plt.title('Filtered Channels Data')
plt.show()
```
